```python
# Standard data libraries
import os
import pandas as pd
import numpy as np

# Plotting
import matplotlib.pyplot as plt

# Scikit-learn: model building and evaluation
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    confusion_matrix, classification_report, roc_curve
)
import joblib  # to save models
import xgboost as xgb  # eXtreme Gradient Boosting

RANDOM_STATE = 42
print('Imports OK â€” ready to go!')

# Try common locations for the CSV
possible_paths = [
    "water_potability.csv",
    "/content/water_potability.csv",
    "/mnt/data/water_potability.csv"
]

csv_path = None
for p in possible_paths:
    if os.path.exists(p):
        csv_path = p
        break

if csv_path is None:
    print("File not found in common paths. Please upload 'water_potability.csv' to Colab (Files pane) or place it next to this notebook and re-run.")
else:
    print("Found CSV at:", csv_path)
    df = pd.read_csv(csv_path)
    print("Shape:", df.shape)
    display(df.head())



# Basic info and missing values
print('\n--- Data Info ---\n')
display(df.info())

print('\n--- Descriptive Stats ---\n')
display(df.describe())

print('\n--- Missing Values per column ---\n')
print(df.isna().sum())

print('\n--- Potability class counts ---\n')
print(df['Potability'].value_counts(dropna=False))
print('\nClass distribution (percent):\n', df['Potability'].value_counts(normalize=True)*100)


# Create X and y
X = df.drop(columns=['Potability'])
y = df['Potability'].astype(int)

# Quick check
print("Feature columns:", list(X.columns))
print("X shape:", X.shape, "y shape:", y.shape)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Train class balance:\n", y_train.value_counts(normalize=True))
print("Test class balance:\n", y_test.value_counts(normalize=True))

# Logistic Regression pipeline (impute + scale + model)
pipe_lr = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=500, class_weight='balanced', random_state=RANDOM_STATE))
])

# Random Forest pipeline (impute + model)
pipe_rf = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('model', RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE))
])

# XGBoost model (we'll create pipeline for consistency)
pipe_xgb = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('model', xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', n_estimators=100, random_state=RANDOM_STATE))
])

# Fit quick versions
print("Training quick baseline models (this should be fast)...")
pipe_lr.fit(X_train, y_train)
pipe_rf.fit(X_train, y_train)
pipe_xgb.fit(X_train, y_train)

# Evaluate quick baseline
def get_metrics(name, model, X_test, y_test, needs_scaling=False):
    if needs_scaling:
        X_eval = StandardScaler().fit_transform(X_test)  # simple scaling for quick eval
    else:
        X_eval = X_test
    y_pred = model.predict(X_eval) if not hasattr(model, 'predict_proba') else model.predict(X_eval)
    if hasattr(model, 'predict_proba'):
        y_prob = model.predict_proba(X_eval)[:,1]
    else:
        y_prob = None
    metrics = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, zero_division=0),
        'Recall': recall_score(y_test, y_pred, zero_division=0),
        'F1': f1_score(y_test, y_pred, zero_division=0),
    }
    if y_prob is not None:
        metrics['ROC-AUC'] = roc_auc_score(y_test, y_prob)
    return metrics

results_quick = {
    'Logistic Regression': get_metrics('LR', pipe_lr, X_test, y_test, needs_scaling=False),
    'Random Forest': get_metrics('RF', pipe_rf, X_test, y_test, needs_scaling=False),
    'XGBoost': get_metrics('XGB', pipe_xgb, X_test, y_test, needs_scaling=False)
}

import pandas as pd
display(pd.DataFrame(results_quick).T)
# Prepare parameter distributions for RandomizedSearchCV - keep these small for speed
from scipy.stats import randint

rf_param_dist = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [None, 8, 12],
    'model__min_samples_split': [2, 5, 10]
}

rf_search = RandomizedSearchCV(
    estimator=pipe_rf,
    param_distributions=rf_param_dist,
    n_iter=5,  # try 5 combinations (keep small to save time)
    scoring='f1',
    cv=3,
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbose=1
)

print("Starting Random Forest hyperparameter search (this may take a few minutes)...")
rf_search.fit(X_train, y_train)
print("Best RF params:", rf_search.best_params_)
best_rf = rf_search.best_estimator_

# XGBoost hyperparameter distribution
xgb_param_dist = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [3, 5, 7],
    'model__learning_rate': [0.01, 0.1, 0.2]
}

xgb_search = RandomizedSearchCV(
    estimator=pipe_xgb,
    param_distributions=xgb_param_dist,
    n_iter=5,
    scoring='f1',
    cv=3,
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbose=1
)

print("Starting XGBoost hyperparameter search (this may take a few minutes)...")
xgb_search.fit(X_train, y_train)
print("Best XGB params:", xgb_search.best_params_)
best_xgb = xgb_search.best_estimator_

lr_param_grid = {'model__C': [0.01, 0.1, 1, 10]}
lr_search = GridSearchCV(pipe_lr, param_grid=lr_param_grid, scoring='f1', cv=3, n_jobs=-1, verbose=1)
print("Starting Logistic Regression grid search (quick)...")
lr_search.fit(X_train, y_train)
print("Best LR params:", lr_search.best_params_)
best_lr = lr_search.best_estimator_


models = {
    'Logistic Regression (tuned)': best_lr if 'best_lr' in globals() else pipe_lr,
    'Random Forest (tuned)': best_rf,
    'XGBoost (tuned)': best_xgb
}

def evaluate_all(models_dict, X_test, y_test):
    rows = []
    for name, model in models_dict.items():
        # If model pipeline expects raw data (we used pipelines), just pass X_test
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:,1] if hasattr(model, 'predict_proba') else None

        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, zero_division=0)
        rec = recall_score(y_test, y_pred, zero_division=0)
        f1 = f1_score(y_test, y_pred, zero_division=0)
        roc = roc_auc_score(y_test, y_prob) if y_prob is not None else np.nan

        rows.append({
            'Model': name,
            'Accuracy': acc,
            'Precision': prec,
            'Recall': rec,
            'F1': f1,
            'ROC-AUC': roc
        })
    return pd.DataFrame(rows).sort_values('F1', ascending=False).reset_index(drop=True)

comparison_df = evaluate_all(models, X_test, y_test)
display(comparison_df)


# Select best model by F1
best_name = comparison_df.iloc[0]['Model']
print("Best model:", best_name)
best_model = models[best_name]

# Confusion matrix
cm = confusion_matrix(y_test, best_model.predict(X_test))
plt.figure(figsize=(5,4))
plt.imshow(cm, cmap='Blues')
plt.title(f'Confusion Matrix: {best_name}')
plt.xlabel('Predicted')
plt.ylabel('Actual')
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i,j], ha='center', va='center', color='red', fontsize=14)
plt.show()

# ROC curve
plt.figure(figsize=(6,5))
for name, model in models.items():
    if hasattr(model, 'predict_proba'):
        fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test)[:,1])
        plt.plot(fpr, tpr, label=name)
plt.plot([0,1],[0,1], 'k--')
plt.title('ROC Curves')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# Random Forest importances (if available)
if 'RandomForest' in str(type(best_rf.named_steps['model'])) or hasattr(best_rf.named_steps['model'], 'feature_importances_'):
    rf_feat = best_rf.named_steps['model'].feature_importances_
    rf_df = pd.DataFrame({'feature': X.columns, 'importance': rf_feat}).sort_values('importance', ascending=False)
    display(rf_df)
    plt.figure(figsize=(8,4))
    plt.bar(rf_df['feature'], rf_df['importance'])
    plt.xticks(rotation=45, ha='right')
    plt.title('Random Forest Feature Importances')
    plt.tight_layout()
    plt.show()
else:
    print('Random Forest importances not available.')

# XGBoost feature importances
if hasattr(best_xgb.named_steps['model'], 'feature_importances_'):
    xgb_feat = best_xgb.named_steps['model'].feature_importances_
    xgb_df = pd.DataFrame({'feature': X.columns, 'importance': xgb_feat}).sort_values('importance', ascending=False)
    display(xgb_df)
    plt.figure(figsize=(8,4))
    plt.bar(xgb_df['feature'], xgb_df['importance'])
    plt.xticks(rotation=45, ha='right')
    plt.title('XGBoost Feature Importances')
    plt.tight_layout()
    plt.show()
else:
    print('XGBoost importances not available.')


# Save the best model (by F1)
filename = 'best_model_week3.joblib'
joblib.dump(best_model, filename)
print("Saved best model to", filename)
```
